```python
import numpy as np
from numpy import nan as NA
from queue import PriorityQueue
```

```python
class stack(list):
    def __init__(self):
        self.items = []
    
    def push(self, item):
        self.items.append(item)
    
    def pop(self):
        if self.is_empty():
            return -1
        return self.items.pop()
    
    def is_empty(self):
        if len(self.items) == 0:
            return True
        return False
```
